4. **stats** - A dictionary of statistics that you will be calculating from the input. (dictionary)

```python
class Statistics_data:

  def __init__(self,array):
    
    self.data = array 

    min,max= np.min(self.data), np.max(self.data)
    n_arr=[]

    for i in array:
      temp= ((i-min)/(max-min))*100
      n_arr.append(temp)


    self.data=n_arr

    self.num_of_stu=0
    self.name_of_teach = "Yourname"
    self.stats = {}
    self.summary=[]

  def set_num_students(self):
    self.num_of_stu= len(self.data)
  def statistics_calculator(self):
    self.stats = {'mean': np.mean(self.data), 'median' : np.median(self.data),
                  'maximum': np.max(self.data), 'minimum' : np.mean(self.data),
                  'std deviation' : np.std(self.data)}
  def organiser(self):
    self.summary = [self.num_of_stu, self.name_of_teach , self.summary]
    print (self.summary, sep='\n')
tst = Statistics_data(data) 
tst.organiser() 
```
